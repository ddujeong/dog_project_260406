```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/aihub_body/detailed_body_dataset.csv")

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False


plt.figure(figsize=(6,4))
sns.countplot(x="detailed_body_label", data=df, order=["slim","standard","broad","overweight"])
plt.title("Body Type 분포")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="detailed_body_label", y="weight_height_ratio", data=df)
plt.title("Weight/Height Ratio 체형분류 기준")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

cm = np.array([
    [7, 1, 0, 0],
    [7, 80, 16, 21],
    [2, 27, 59, 12],
    [1, 2, 0, 73]
])

labels = ["slim", "standard", "broad", "overweight"]

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=labels, yticklabels=labels)

plt.title("Confusion Matrix (Body Type)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["slim", "standard", "broad", "overweight"]
f1_scores = [0.56, 0.68, 0.67, 0.80]

plt.figure(figsize=(6,4))
plt.bar(labels, f1_scores)

plt.title("F1-score by Body Type")
plt.ylim(0,1)
plt.ylabel("F1-score")

plt.tight_layout()
plt.show()
```

```python
recall = [0.88, 0.65, 0.59, 0.96]

plt.figure(figsize=(6,4))
plt.bar(labels, recall)

plt.title("Recall by Body Type")
plt.ylim(0,1)
plt.ylabel("Recall")

plt.tight_layout()
plt.show()
```
