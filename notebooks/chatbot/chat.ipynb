```python
import os
import json
import numpy as np
from sentence_transformers import SentenceTransformer

data_path = 'data/aihub_chat/processed/merged_dataset.json'
save_path = 'data/aihub_chat/processed/embeddings.npy'

print(f"현재 경로: {os.getcwd()}")
```

```python
# 1. 데이터 로드
with open(data_path, "r", encoding="utf-8") as f:
    docs = json.load(f)

doc_texts = [d["content"] for d in docs]
print(f"총 데이터 개수: {len(doc_texts)}개")

# 2. 모델 로드 (서버와 동일한 모델 사용)
model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)
print(f"모델 '{model_name}' 로드 완료.")
```

```python
print("🚀 임베딩 생성을 시작합니다. (916배치 예상)")

# 배차 사이즈는 32로 설정 (로그와 동일하게)
doc_embeddings = model.encode(
    doc_texts, 
    batch_size=32, 
    show_progress_bar=True
)

print("\n✅ 임베딩 생성 완료!")
print(f"생성된 벡터 형태: {doc_embeddings.shape}") # (데이터개수, 384) 예상
```

```python
# 3. npy 파일로 저장
np.save(save_path, doc_embeddings)

# 잘 저장됐는지 다시 로드해서 확인
loaded_embeddings = np.load(save_path)
print(f"💾 파일 저장 위치: {save_path}")
print(f"🔍 로드 테스트 결과: {loaded_embeddings.shape} (정상)")

if np.array_equal(doc_embeddings, loaded_embeddings):
    print("✨ 원본 데이터와 저장된 데이터가 일치합니다. 이제 Push 하셔도 됩니다!")
```
